### Restaurant Data Sorting Architecture: DynamoDB vs Bedrock KB

For this Restaurant Assistant example, here’s a clear separation of what data should be stored in **Amazon DynamoDB** versus what should be ingested into the **Amazon Bedrock Knowledge Base**, based on the architecture and the description in the notebook.

---

#### Data to Store in DynamoDB

**Amazon DynamoDB** is used for transactional, structured, and frequently changing data—primarily for booking management. You should store:

- **Booking/Reservation Records**
  - `booking_id`: Unique identifier for each reservation  
  - `restaurant_name`: Name of the restaurant  
  - `date`: Reservation date (YYYY-MM-DD)  
  - `hour`: Reservation time (HH:MM)  
  - `guest_name`: Name of the person making the reservation  
  - `num_guests`: Number of guests for the booking  
- **Other transactional details**
  - Status (e.g., confirmed, canceled, pending)  
  - Contact info for the booking  
  - Special requests (dietary restrictions, seating preferences, etc.)  
- **Audit or activity logs** (optional)
  - Track changes or actions for compliance and debugging  

**Purpose:**  
DynamoDB is the source of truth for all live bookings and reservation details. It supports creation, retrieval, updating, and deletion of reservations in real time.

---

#### Data to Ingest into the Knowledge Base

The **Amazon Bedrock Knowledge Base** is meant for reference, informational, and relatively static data that the agent uses to answer user questions about restaurants and their menus. You should ingest:

- **Restaurant Directory**
  - List of all restaurants in your system  
  - Restaurant names, addresses, contact details, opening hours  
- **Menus**
  - Menu items (names, descriptions, prices)  
  - Special offers, dietary information, etc.  
- **General Information**
  - FAQs about restaurants (policies, parking info, etc.)  
  - Reviews, ratings, ambience descriptions  
- **Location-specific Info**
  - Maps, directions, nearby attractions  

**Purpose:**  
The Knowledge Base is used for retrieval-augmented generation. When a user asks questions such as “Where can I eat in San Francisco?” or “Show me the menu for Rice & Spice,” the agent fetches this information from the knowledge base.

---

#### Summary Table

| Data Type                     | DynamoDB | Knowledge Base |
|-------------------------------|:--------:|:--------------:|
| Bookings/Reservations         | Yes      | No             |
| Restaurant Directory          | No       | Yes            |
| Menus                         | No       | Yes            |
| Booking Details               | Yes      | No             |
| General Restaurant Info/FAQs  | No       | Yes            |
| Audit Logs (optional)         | Yes      | No             |
| Reviews, Ratings              | No       | Yes            |

---

**In short:**

- **DynamoDB:** Live, transactional data (bookings, reservations).  
- **Knowledge Base:** Static/reference data (restaurant info, menus, FAQs).
